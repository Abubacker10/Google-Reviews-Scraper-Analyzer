```python
from transformers import pipeline
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
def model():
    pipe = pipeline("text-classification", model="MarieAngeA13/Sentiment-Analysis-BERT")
    return pipe
```

```python
pipe = pipeline("text-classification", model="MarieAngeA13/Sentiment-Analysis-BERT")
pipe('boring')[0]
```

```python
def sentiment(i):
    pipe = model()
    if isinstance(i, str) and len(i) > 10:
        res = pipe(i)[0]
        return res['label'], res['score']
    else:
        return None,None
```

```python
def analysis():
    data = pd.read_csv('reviews_texts.csv',header=None)
    data.columns=['Reviews']
    data['Label'],data['Score']  = np.vectorize(sentiment)(data['Reviews'])
    data.to_csv('Google_reviews_analysis.csv',index=False)
```
